```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(-3,3,01.1)
y1=np.sin(x)
y2=np.cos(x)
```

```python
%matplotlib inline
plt.figure()
plt.plot(x,y1,'ro-',label='sin')
plt.plot(x,y2,'b--',label='cos')
plt.legend()
plt.title('aa')
ax=plt.gca()
```
